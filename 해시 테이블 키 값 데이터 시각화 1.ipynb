```python
import matplotlib.pyplot as plt

# 해시 테이블 상태 시뮬레이션
indices = list(range(13))
keys = [50, None, None, 55, None, 18, None, None, None, 22, 35, 37, 25]
data = ['orange', None, None, 'cherry', None, 'banana', None, None, None,
        'mango', 'lime', 'apple', 'grape']

# 시각화를 위한 색상
colors = ['lightgray' if k is None else 'skyblue' for k in keys]

fig, ax = plt.subplots(figsize=(12, 2))
bars = ax.bar(indices, [1]*13, tick_label=indices, color=colors, edgecolor='black')

# 키 라벨
for idx, (k, d) in enumerate(zip(keys, data)):
    label = f"{k}\n{d}" if k is not None else ""
    ax.text(idx, 0.5, label, ha='center', va='center', fontsize=8)

ax.set_ylim(0, 1.5)
ax.set_title("Linear Probing", fontsize=14)
ax.axis('off')

plt.tight_layout()
plt.show()
```
